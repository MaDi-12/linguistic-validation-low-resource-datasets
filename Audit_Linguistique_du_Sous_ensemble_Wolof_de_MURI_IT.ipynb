<h2 id="methodologie" style="color:#004080; border-bottom:3px solid #004080; padding-bottom:5px;">
Cadre méthodologique
</h2>

<p>
L’évaluation repose sur une stratégie hiérarchique de détection linguistique
visant à tester la robustesse du signal wolof dans un espace linguistique ouvert,
puis dans un espace spécialisé.
</p>

<p>
L’utilisation exclusive d’un détecteur spécialisé pour les langues africaines
n’est pas suffisante dans ce contexte. En effet, ces modèles opèrent dans un
espace restreint et ne couvrent pas l’ensemble des langues globalement dominantes,
notamment l’anglais et le français. Lorsqu’un texte appartient à une langue
non couverte par le modèle, celui-ci peut produire une attribution contrainte
vers la classe la moins improbable, sans que cela ne reflète un signal linguistique réel.
</p>

<p>
Afin d’éviter ce biais structurel, l’analyse procède en trois étapes :
</p>

<ul>
<li>
<strong>Étape 1 — Filtrage multilingue global :</strong>
Application d’un classifieur couvrant un large ensemble de langues,
afin d’identifier les langues dominantes dans un espace ouvert
et de détecter la présence éventuelle de langues non africaines.
</li>

<li>
<strong>Étape 2 — Calibration par seuil de confiance :</strong>
Séparation des prédictions robustes des cas ambigus,
sur la base des niveaux de confiance fournis par le classifieur global.
</li>

<li>
<strong>Étape 3 — Analyse spécialisée africaine :</strong>
Application d’un détecteur spécialisé sur les cas résiduels,
afin d’examiner la présence d’un signal wolof dans un espace linguistique restreint.
</li>
</ul>

<p>
Cette approche hiérarchique permet de distinguer les langues
clairement identifiables à l’échelle globale des cas nécessitant
une analyse intra-africaine plus fine,
et d’évaluer la détectabilité effective du signal wolof
dans le sous-ensemble étudié.
</p>

```python
import fasttext

ft_model = fasttext.load_model("lid.176.bin")
```

```python
# ===============================
# Détection de langue avec fastText
# ===============================

from tqdm import tqdm
import pandas as pd

def detect_language_fasttext(text):
    try:
        labels, probabilities = ft_model.predict(str(text).replace("\n", " "), k=1)

        language = labels[0].replace("__label__", "")
        confidence = float(probabilities[0])

        return language, confidence

    except Exception as e:
        return "error", 0.0
```

```python
from tqdm import tqdm
import pandas as pd

tqdm.pandas()

df_wolof[["ft_pred_lang", "ft_confidence"]] = (
    df_wolof["input"]
    .progress_apply(lambda x: pd.Series(detect_language_fasttext(x)))
)
```

```python
# ===============================
# Résumé statistique — fastText (Instructions)
# ===============================

total = len(df_wolof)

# Distribution complète des langues
language_counts = df_wolof["ft_pred_lang"].value_counts()
language_percentages = df_wolof["ft_pred_lang"].value_counts(normalize=True) * 100

summary_df = pd.DataFrame({
    "Nombre": language_counts,
    "Pourcentage (%)": language_percentages.round(2)
})

print("📊 Distribution complète des langues détectées (fastText)")
display(summary_df.head(10))

# Taux de wolof
wolof_detected = language_counts.get("wo", 0)
wolof_rate = wolof_detected / total

print("\n" + "-"*50)
print("📌 Indicateurs clés")
print(f"Nombre total d’instructions analysées : {total}")
print(f"Nombre détecté comme wolof : {wolof_detected}")
print(f"Taux de wolof détecté : {wolof_rate:.2%}")
print(f"Taux estimé de non-wolof : {(1 - wolof_rate):.2%}")

# Analyse des scores de confiance
print("\n" + "-"*50)
print("📈 Statistiques des scores de confiance")
display(df_wolof["ft_confidence"].describe())
```

```python
# ===============================
# Séparation par seuil de confiance (fastText)
# ===============================

CONFIDENCE_THRESHOLD = 0.7

# Marquage haute / basse confiance
df_wolof["ft_high_confidence"] = df_wolof["ft_confidence"] >= CONFIDENCE_THRESHOLD

# Séparation des sous-ensembles
df_high_conf = df_wolof[df_wolof["ft_high_confidence"]].copy()
df_low_conf = df_wolof[~df_wolof["ft_high_confidence"]].copy()

# Statistiques
total = len(df_wolof)
high_count = len(df_high_conf)
low_count = len(df_low_conf)

print("📊 Résumé de la séparation par seuil")
print("-" * 50)
print(f"Seuil de confiance τ = {CONFIDENCE_THRESHOLD}")
print(f"Total instructions : {total}")
print(f"Haute confiance (≥ τ) : {high_count} ({high_count/total:.2%})")
print(f"Cas ambigus (< τ) : {low_count} ({low_count/total:.2%})")
```

```python
from transformers import pipeline
from tqdm import tqdm
tqdm.pandas()

classifier_afrolid = pipeline(
    "text-classification",
    model="UBC-NLP/afrolid_1.5",
    return_all_scores=True
)

print("AfroLID 1.5 chargé depuis HuggingFace.")
```

```python
# ============================================================
# Fonction de prédiction robuste avec gestion des textes longs
# Agrégation : max score parmi les segments
# ============================================================

def split_text(text, max_chars=300):
    """
    Découpe un texte en segments de longueur max_chars.
    """
    return [text[i:i+max_chars] for i in range(0, len(text), max_chars)]


def predict_afrolid(texts, max_chars=300):
    predicted_languages = []
    confidence_scores = []

    for text in tqdm(texts):

        best_label = None
        best_score = 0

        if len(text) <= max_chars:

            results = classifier_afrolid(text)[0]

            best_entry = max(results, key=lambda x: x['score'])
            best_label = best_entry['label']
            best_score = best_entry['score']

        else:

            segments = split_text(text, max_chars=max_chars)

            for seg in segments:

                results = classifier_afrolid(seg)[0]

                seg_best = max(results, key=lambda x: x['score'])

                if seg_best['score'] > best_score:
                    best_score = seg_best['score']
                    best_label = seg_best['label']

        predicted_languages.append(best_label)
        confidence_scores.append(best_score)

    return predicted_languages, confidence_scores
```

```python
# ===============================
# Vérification AfroLID sur cas ambigus et analyse des résultats
# ===============================

afrolid_predicted_langs, afrolid_confidence_scores = predict_afrolid(df_low_conf["input"].tolist())

df_low_conf["afrolid_lang"] = afrolid_predicted_langs
df_low_conf["afrolid_conf"] = afrolid_confidence_scores

print("✅ Vérification AfroLID des cas ambigus terminée.")

# ===============================
# Analyse des résultats AfroLID pour les cas ambigus
# ===============================

total_low_conf = len(df_low_conf)

print(f"\n📊 Résumé AfroLID pour les {total_low_conf} cas ambigus (fastText confidence < {CONFIDENCE_THRESHOLD})")
print("-" * 70)


lang_counts_afrolid = df_low_conf["afrolid_lang"].value_counts()
lang_percentages_afrolid = df_low_conf["afrolid_lang"].value_counts(normalize=True) * 100

summary_afrolid_df = pd.DataFrame({
    "Nombre": lang_counts_afrolid,
    "Pourcentage (%)": lang_percentages_afrolid.round(2)
})

print("Distribution des langues détectées par AfroLID :")
display(summary_afrolid_df.head(10))

# Taux de wolof détecté par AfroLID parmi les cas ambigus
# .get() will now work correctly as the index contains strings
wolof_detected_afrolid = lang_counts_afrolid.get("wol", 0)
wolof_rate_afrolid = wolof_detected_afrolid / total_low_conf

print("\n" + "-"*70)
print("📌 Indicateurs clés (AfroLID sur cas ambigus)")
print(f"Instructions détectées comme Wolof par AfroLID : {wolof_detected_afrolid}")
print(f"Taux de Wolof détecté par AfroLID : {wolof_rate_afrolid:.2%}")

# Analyse des scores de confiance d'AfroLID
print("\n" + "-"*70)
print("📈 Statistiques des scores de confiance AfroLID")
display(df_low_conf["afrolid_conf"].describe())

# Impact global estimé
total_global = len(df_wolof)
# We sum up the high confidence wolof detections from fastText (if any) and the low confidence wolof detections from AfroLID
# Assuming df_high_conf was already processed and we know its wolof count
# As per earlier FastText results, wolof_detected for fasttext was 0. So, we only consider wolof_detected_afrolid here.
wolof_global_estimated = wolof_detected_afrolid # Since fastText found 0 wolof at high confidence
wolof_global_rate = wolof_global_estimated / total_global

print("\n" + "-"*70)
print("🌍 Estimation globale après triangulation")
print(f"Wolof détecté (global) : {wolof_global_estimated}")
print(f"Taux global estimé de wolof : {wolof_global_rate:.2%}")
```
